```python
from nsetools import Nse
from pprint import pprint
import time
```

```python
nse = Nse()
code_list = nse.get_stock_codes() #Add progress indicator
```

```python
meta_data = []
exception_tickers = []
for c in list(code_list.keys()):
    
    try:
        data = nse.get_quote(c)
        meta_data.append(data)
    
    except Exception as e:
        print("Exception {} occured for ticker: {}".format(e, c))
        exception_tickers.append(c)
        
    time.sleep(10)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(meta_data, orient='columns')
```
